```python
# 必要なライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


# Van der Pol方程式の定義
def van_der_pol(t, z, mu):
    x, y = z
    dxdt = y
    dydt = mu * (1 - x**2) * y - x
    return [dxdt, dydt]


# パラメータ設定
mu = 0.1  # muの値を変更して異なるベクトル場を得ることができます
t_span = (0, 20)
t_eval = np.linspace(*t_span, 1000)  # 評価ポイントの数を増やす
initial_conditions = [1.0, 0.0]  # 初期条件 [x0, y0]

# Van der Pol方程式の数値解を求める
sol = solve_ivp(van_der_pol, t_span, initial_conditions, args=(mu,), t_eval=t_eval)

# ベクトル場の格子ポイントを作成
x_range = np.linspace(-2, 2, 20)
y_range = np.linspace(-1.5, 1.5, 20)
x, y = np.meshgrid(x_range, y_range)

# ベクトル場を計算
dxdt = y
dydt = mu * (1 - x**2) * y - x

# ベクトル場の矢印をプロット
plt.figure(figsize=(8, 6))
plt.quiver(x, y, dxdt, dydt, scale=20)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Van der Pol Equation Vector Field (mu = {mu})")
plt.grid(True)

# 解軌道の線をプロット
plt.plot(
    sol.y[0],
    sol.y[1],
    label="Van der Pol Equation Trajectory",
    color="red",
    linewidth=2,
)

plt.legend()
plt.show()
```
