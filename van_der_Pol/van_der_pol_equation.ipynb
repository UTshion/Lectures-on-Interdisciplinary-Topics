```python
# 
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


# Van der Pol
def van_der_pol(t, z, mu):
    x, y = z
    dxdt = y
    dydt = mu * (1 - x**2) * y - x
    return [dxdt, dydt]


# 
mu = 0.1  # mu
t_span = (0, 20)
t_eval = np.linspace(*t_span, 1000)  # 
initial_conditions = [1.0, 0.0]  #  [x0, y0]

# Van der Pol
sol = solve_ivp(van_der_pol, t_span, initial_conditions, args=(mu,), t_eval=t_eval)

# 
x_range = np.linspace(-2, 2, 20)
y_range = np.linspace(-1.5, 1.5, 20)
x, y = np.meshgrid(x_range, y_range)

# 
dxdt = y
dydt = mu * (1 - x**2) * y - x

# 
plt.figure(figsize=(8, 6))
plt.quiver(x, y, dxdt, dydt, scale=20)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Van der Pol Equation Vector Field (mu = {mu})")
plt.grid(True)

# 
plt.plot(
    sol.y[0],
    sol.y[1],
    label="Van der Pol Equation Trajectory",
    color="red",
    linewidth=2,
)

plt.legend()
plt.show()
```
